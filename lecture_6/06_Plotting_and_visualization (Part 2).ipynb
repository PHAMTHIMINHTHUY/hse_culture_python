### 3. Line plots

Посмотрим, как менялась длина постов с течением времени.
Для этого нам понадобится файл с текстами постов. Если у вас его нет, можете использовать `texts.txt` из этой же папки.

```python
post_lengths = []

with open("texts.txt", 'r', encoding='utf-8') as f:
    for line in f:
        post_lengths.append(len(line.strip()))
```

```python
post_lengths = list(reversed(post_lengths))
```

Зачем мы развернули посты? (они теперь в обратном порядке)

```python
plt.figure(figsize=(10,5)) 
plt.plot(post_lengths)
plt.ylabel('post lengths in symbols')
plt.xlabel('post number')
plt.show()
```

А это те же данные, но со сглаживанием (чтобы проследить тенденцию)

```python
from scipy.ndimage.filters import gaussian_filter1d

ysmoothed = gaussian_filter1d(post_lengths, sigma=10)
plt.plot(ysmoothed)
plt.show()
```

А как его сохранить?

```python
plt.figure(figsize=(10,5)) 
plt.plot(ysmoothed)
plt.savefig("my_plot.png", dpi=1000) # dpi это что-то вроде качества изображения
```

### 4. Scatter plot

```python
from nltk.tokenize import word_tokenize
from collections import Counter
```

```python
with open("texts.txt", 'r', encoding='utf-8') as f:
    texts = f.read().lower()
```

```python
words = word_tokenize(texts)
len(words)
```

```python
words = [i for i in words if i.isalnum()]
len(words)
```

```python
counted_words = dict(Counter(words).most_common(100))
```

```python
plt.figure(figsize=(20,10)) 
plt.scatter(counted_words.keys(), counted_words.values())
plt.ylabel('Word count')
plt.xlabel('Word rank')
plt.xticks(range(100), range(100))
plt.show()
```

Мы с вами дружно и вместе проверили закон Ципфа. 


**Википедия:** Зако́н Ци́пфа («ранг—частота») — эмпирическая закономерность распределения частоты слов естественного языка: если все слова языка (или просто достаточно длинного текста) упорядочить по убыванию частоты их использования, то частота n-го слова в таком списке окажется приблизительно обратно пропорциональной его порядковому номеру n (так называемому рангу этого слова, см. шкала порядка). Например, второе по используемости слово встречается примерно в два раза реже, чем первое, третье — в три раза реже, чем первое, и так далее.
